```python
import os
judges = os.listdir()
judges.remove(".ipynb_checkpoints")
judges.remove("script_count.ipynb")

t_attempted = 0
t_solved = 0
t_sub = 0
t_ac_sub = 0
t_line = 0

output = ""
markdown = ""

for judge in judges:
    
    attempted = 0
    solved = 0
    sub = 0
    ac_sub = 0
    line = 0
    
    output += judge
    markdown += "|"+judge
    output += "\n"
    
    problems = os.listdir(judge+"/")
    for problem in problems:
        attempted += 1
        
        submissions = os.listdir(judge+"/"+problem+"/")
        ac = 0
        
        for submission in submissions:
            sub += 1
            
            if ("Accepted" in submission) or ("AC" in submission):
                ac_sub += 1
                ac = 1
                
            with open(judge+"/"+problem+"/"+submission, "r") as f:
                for s in f:
                    line += 1
                f.close()
            line -= 6
        solved += ac
    now = ""
    now += f"Attempted: {attempted} \n"
    now += f"Solved: {solved} \n"
    now += f"Submission Made: {sub} \n"
    now += f"Correct Submission: {ac_sub} \n"
    now += f"Lines of Code: {line} \n"
    now += "\n\n"
    output += now
    
    markdown += "|"+str(attempted)
    markdown += "|"+str(solved)
    markdown += "|"+str(sub)
    markdown += "|"+str(ac_sub)
    markdown += "|"
    markdown += "\n"
    
    #os.remove(f"C:/Users/Araf/Desktop/Github/Problem-Solving/{judge}/README.md.txt")
    #with open(f"C:/Users/Araf/Desktop/Github/Problem-Solving/{judge}/README.md", "w") as f:
        #f.write(now)
        #f.close()
    
    t_attempted += attempted
    t_solved += solved
    t_sub += sub
    t_ac_sub += ac_sub
    t_line += line

output += "Total\n"
output += f"t_attempted: {t_attempted} \n"
output += f"t_solved: {t_solved} \n"
output += f"t_sub: {t_sub} \n"
output += f"t_ac_sub: {t_ac_sub} \n"
output += f"t_line: {t_line} \n"
output += "\n\n"

print(markdown)
```
