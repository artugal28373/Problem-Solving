```python
import os
path = 'codeforces/'
path_sorted = "codeforces_sorted/"
files = os.listdir(path)


def get_verdict(file):
    verdict = ""
    with open(f"{path}{file}", 'r') as f:
        s = f.read()
        ii = s.find('solution_verdict: ')+len('solution_verdict: ')
        
        pr = " "
        while ii<len(s):
            if s[ii]==' ' and pr==' ':
                break
            verdict += (s[ii])
            pr = s[ii]
            ii += 1
    f.close()     
    return verdict.strip()

def get_name(file):
    ii = file.find('(')
    file = file[ii+1:-5].strip()
    file = file.replace('.', 'x')
    return file

for file in files:
    #print(get_verdict(file))
    #print(get_name(file))
    
    verdict = get_verdict(file)
    name = get_name(file)
    
    current = path_sorted+name
    
    if not os.path.exists(current):
        os.mkdir(current)
    
    with open(f"{path}{file}", 'r') as f:
        code = f.read()
    f.close()
    #os.remove(f"{path}{file}")
    
    current = path_sorted+name+f'/{verdict}'
    for j in range(1, 1000):
        if (j==1):
            now = current+'.cpp'
        else:
            now = current+f' ({j}).cpp'

        if not os.path.exists(now):
            with open(now, 'w') as f:
                f.write(code)
            f.close()

            break
    
```
