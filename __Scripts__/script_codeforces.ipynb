```python
import os
import time

from selenium import webdriver
from selenium.webdriver.common.by import By

chromedriver_path = "C:\\Users\\Araf\\Desktop\\Script\\chromedriver_win32"
os.environ['PATH'] += chromedriver_path
```

```python
driver = webdriver.Chrome()
driver.implicitly_wait(40)

driver.get("https://codeforces.com/enter")
```

```python
login = driver.find_element(By.ID, "handleOrEmail")
password = driver.find_element(By.ID, "password")

handle = 'kzvd4729'
_pass = 'earth_is_flat'

login.send_keys(handle)
password.send_keys(_pass)
time.sleep(3)
btn = driver.find_element(By.CSS_SELECTOR, ".submit")
btn.click()
time.sleep(3)
```

```python
lt = 1
rt = 145
for i in range(lt, rt+1):  
    driver.get(f"https://codeforces.com/submissions/{handle}/page/{i}")
    time.sleep(7)
    count = 0
    
    table_row = driver.find_elements(By.TAG_NAME, "tr") # every row is selected
    table_row = table_row[1:-2]
    
    ii = 0
    t_len = len(table_row)
    while ii<t_len:
        row = table_row[ii]
        table_col = row.find_elements(By.TAG_NAME, "td")

        subid = table_col[0].text
        subid = subid.strip()
        
        author = table_col[2].text
        author = author.strip()

        created = table_col[1].text
        if "UTC" in created:
            ind = created.find("UTC")
            created = created[:ind]
        created = created.strip()

        problem_title = table_col[3].text
        problem_title = problem_title.strip()

        problem_link = table_col[3].find_element(By.TAG_NAME, "a")
        problem_link = problem_link.get_attribute("href")
        problem_link = problem_link.strip()

        language = table_col[4].text
        language = language.strip()

        verdict = table_col[5].text
        verdict = verdict.strip()

        run_time = table_col[6].text
        run_time = run_time.strip()

        mem_used = table_col[7].text
        mem_used = mem_used.strip()

        
        btn = table_col[0].find_element(By.CSS_SELECTOR, ".view-source")
        btn.click()
        time.sleep(2)
        

        table_data=[
            [f'*  @author: {author}', f'created: {created}'],
            [f'*  solution_verdict: {verdict}', f'language: {language}'],
            [f'*  run_time: {run_time}', f'memory_used: {mem_used}']
        ]
        printer = '/****************************************************************************************\n'
        for row in table_data:
            printer+=("{: <60} {: <50}".format(*row))+'\n'
        printer += f"*  problem: {problem_link}\n"
        printer +='****************************************************************************************/\n'
        #print(printer)
        
        
        code = driver.find_elements(By.TAG_NAME, "code")  
        for line in code:
            printer+=(line.text)
        
        
        # insurance
        if len(printer)<590:
            time.sleep(3)
            print("got one")
            close = driver.find_element(By.CSS_SELECTOR, ".close")
            close.click()
            time.sleep(2)
            continue
        
            
        
        for c in "\/:*?\"<>|": # folder name can't have these character
            problem_title = problem_title.replace(c, 'x')
        path = f'codeforces/{subid}({problem_title}).cpp'

        with open(path, 'w') as f:
            f.write(printer)
        
        
        close = driver.find_element(By.CSS_SELECTOR, ".close")
        close.click()
        
        
        count += 1
        print(f"page {i} sub {count} len {len(printer)}")
        time.sleep(1)
        ii += 1
```
