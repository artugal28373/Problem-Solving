```python
import os
import time

from selenium import webdriver
from selenium.webdriver.common.by import By

chromedriver_path = "C:\\Users\\Araf\\Desktop\\Script\\chromedriver_win32"
os.environ['PATH'] += chromedriver_path
```

```python
driver = webdriver.Chrome()
driver.implicitly_wait(40)

driver.get("https://leetcode.com/accounts/login/")
# do login manually
```

```python
problem_list = []
def get_problem_list():
    # there are total 14 submission page
    for page in range(1,15):
        driver.get(f"https://leetcode.com/submissions/#/{page}")
        time.sleep(5)
        
        table_row = driver.find_elements(By.TAG_NAME, "tr")
        table_row = table_row[1:]
        
        for row in table_row:
            table_col = row.find_elements(By.TAG_NAME, "td")
            problem_list.append(table_col[1].text)

get_problem_list()
problem_list = list(set(problem_list))
```

```python
cnt = 0
for p in problem_list:
    current = f"leetcode/{p}"
    if not os.path.exists(current):
        cnt+=1
print(cnt)
```

```python
for i in range(86, 100):
    print(problem_list[i])
```

```python
# there are 100 problem in problem_list
lt=0
rt=99

for prb in range(lt, rt+1):
    
    p_title = problem_list[prb].strip()
    
    ck = f"leetcode/{p_title}"
    if os.path.exists(ck):
        continue
                      
    
    ln = problem_list[prb].lower()

    ln = ln.replace('-', ' ')
    ln = ln.replace('  ', ' ')
    ln = ln.replace('  ', ' ')
    
    ln = ln.replace(' ', '-')
    ln = ln.replace('\'', '')
    ln = ln.replace('(', '')
    ln = ln.replace(')', '')
    

    p_link = f"https://leetcode.com/problems/{ln}"
    driver.get(p_link)
    time.sleep(7)

    sub_btn = driver.find_element(By.CSS_SELECTOR, "a[href*='submission']")
    sub_btn.click()
    time.sleep(2)
    
    
    t_row = driver.find_elements(By.TAG_NAME, "tr")
    t_row = t_row[1:]
    
    created = []
    verdict = []
    run_time = []
    mem_used = []
    language = []
    sub_link = []
    
    for row in t_row:
        t_col = row.find_elements(By.TAG_NAME, "td")
        created.append(t_col[0].text.strip())
        verdict.append(t_col[1].text.strip())
        run_time.append(t_col[2].text.strip())
        mem_used.append(t_col[3].text.strip())
        language.append(t_col[4].text.strip())
        sub_link.append((t_col[1].find_element(By.CSS_SELECTOR, "a[href*='detail']")).get_attribute("href"))
    
    for i in range(len(sub_link)):
        driver.get(sub_link[i])
        time.sleep(7)
        code = driver.find_element(By.CSS_SELECTOR, ".ace_content").text
        
        
        table_data=[
            [f'*  @author: kzvd4729', f'created: {created[i]}'],
            [f'*  solution_verdict: {verdict[i]}', f'language: {language[i]}'],
            [f'*  run_time: {run_time[i]}', f'memory_used: {mem_used[i]}']
        ]
        printer = '/****************************************************************************************\n'
        for row in table_data:
            printer+=("{: <60} {: <50}".format(*row))+'\n'
        printer += f"*  problem: {p_link}\n"
        printer +='****************************************************************************************/\n'
        
        printer += code
        
        if len(printer)<590:
            print("got one bad")
            exit(0)
        
        for c in "\/:*?\"<>|.": # folder name can't have these character
            p_title = p_title.replace(c, 'x')
        
        current = f"leetcode/{p_title}"
        if not os.path.exists(current):
            os.mkdir(current)
            
        
        current = current+f'/{verdict[i]}'
        for j in range(1, 1000):
            if (j==1):
                now = current+'.cpp'
            else:
                now = current+f' ({j}).cpp'

            if not os.path.exists(now):
                with open(now, 'w') as f:
                    f.write(printer)
                f.close()

                break
        print(f"done {prb}:  {i+1}/{len(sub_link)}   {p_title}")
    
```
