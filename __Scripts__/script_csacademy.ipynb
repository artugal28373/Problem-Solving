```python
import os
import time

from selenium import webdriver
from selenium.webdriver.common.by import By

chromedriver_path = "C:\\Users\\Araf\\Desktop\\Script\\chromedriver_win32"
os.environ['PATH'] += chromedriver_path
```

```python
driver = webdriver.Chrome()
driver.implicitly_wait(40)

driver.get("https://csacademy.com/contest/archive/submissions/")
# do login manually
```

```python
rows = driver.find_elements(By.CSS_SELECTOR, ".submissionSummary-46")
title = []
subid = []

for row in rows:
    txt = row.text
    subid.append(txt[5:12])
    ii = txt.find("Task")
    jj = txt.find(",")
    
    title.append(txt[ii+5:jj])
```

```python
lt=0
rt=len(subid)-1

for i in range(lt, rt+1):
    
    driver.get(f"https://csacademy.com/submission/{subid[i]}/")
    time.sleep(4)
    
    code = driver.find_element(By.CSS_SELECTOR, ".ace_text-layer").text
    
    detail = driver.find_elements(By.XPATH, "//div/div/div/div/div/p")
    verdict = detail[1].text.strip()[9:]
    language = detail[3].text.strip()[10:]
    run_time = detail[4].text.strip()[16:]
    mem_used = detail[5].text.strip()[14:]
    
    
    table_data=[
            [f'*  @author: kzvd4729', f'created: '],
            [f'*  solution_verdict: {verdict}', f'language: {language}'],
            [f'*  run_time: {run_time}', f'memory_used: {mem_used}']
        ]
    printer = '/****************************************************************************************\n'
    for row in table_data:
        printer+=("{: <60} {: <50}".format(*row))+'\n'
    printer += f"*  problem: \n"
    printer +='****************************************************************************************/\n'

    printer += code

    if len(printer)<590:
        print("got one bad")
        break

    #CS Academy

    current = f"CS Academy/{title[i]}"
    if not os.path.exists(current):
        os.mkdir(current)

    current = current+f'/{verdict}'
    for k in range(1, 1000):
        if (k==1):
            now = current+'.cpp'
        else:
            now = current+f' ({k}).cpp'

        if not os.path.exists(now):
            with open(now, 'w') as f:
                f.write(printer)
            f.close()
            break
    print(f"done:  {i}")
```
