```python
sub_links = []
with open("tmp.txt", "r") as f:
    sub_links = list(f.read().split(','))
    f.close()

len(sub_links)
```

```python
# 925
lt = 833
rt = 924
for i in range(lt, rt+1):
    sub = sub_links[i]
    driver.get(sub)
    time.sleep(6)
    
    row = driver.find_element(By.XPATH, '//section/div/ul')
    col = row.find_elements(By.TAG_NAME, 'li')
    
    created = col[0].text.strip()
    run_time = col[2].text.strip()
    mem_used = col[3].text.strip()
    language = driver.find_element(By.XPATH, '//section/aside/div/div/div').text[10:]
    verdict = driver.find_element(By.CSS_SELECTOR, '.solution-status').get_attribute('title')
    
    title = driver.find_element(By.XPATH, "//div/h2/span/strong/a").text.strip()
    p_link = driver.find_element(By.XPATH, "//div/h2/span/strong/a").get_attribute("href")
    
    code = driver.find_element(By.CSS_SELECTOR, '.ace_content').text
    
    subid = sub[-8:]
    
    table_data=[
            [f'*  @author: kzvd4729', f'created: {created}'],
            [f'*  solution_verdict: {verdict}', f'language: {language}'],
            [f'*  run_time: {run_time}', f'memory_used: {mem_used}']
        ]
    printer = '/****************************************************************************************\n'
    for row in table_data:
        printer+=("{: <60} {: <50}".format(*row))+'\n'
    printer += f"*  problem: {p_link}\n"
    printer +='****************************************************************************************/\n'

    printer += code
    
    if len(printer)<590:
        print("got one bad")
        break
    
    name = f"codechef/{subid}){title}){verdict}.cpp"
    with open(name, 'w', encoding='utf8') as f:
        f.write(printer)
        f.close()
    
    print(f"done: {i}")
```
